# Linear Regression

### Fabricate Dataset

Fabricate some data that shows a roughly linear relationship between average vehicle speed, and bumpiness of the road.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

vehicleSpeeds = np.random.normal(40.0, 10.0, 1000)
bumpiness = (1.0 -(vehicleSpeeds / 70.0)) + np.random.normal(0, 0.1, 1000)

sns.scatterplot(vehicleSpeeds, bumpiness)
```

### Create Model and Fit Data

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(vehicleSpeeds, bumpiness)
```

R-squared value shows a decent fit

```python
r_value ** 2
```

### Visualize the Results

Use the slope and intercept got from the regression to plot predicted values vs. observed

```python
def predict(x):
    return slope * x + intercept

sns.set()

fitLine = predict(vehicleSpeeds)

plt.scatter(vehicleSpeeds, bumpiness)
plt.plot(vehicleSpeeds, fitLine, c='r')
plt.show()
```
