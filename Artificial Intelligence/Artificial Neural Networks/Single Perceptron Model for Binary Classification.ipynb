# Single Perceptron Model for Binary Classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from keras.models import Sequential
from keras.layers import Dense

speed_data = pd.read_csv("Speed_Decision_Data.csv")
```

```python
speed_data.head()
```

```python
sns.scatterplot(speed_data['1 - Distance From Bump'], speed_data['Height of Bump'], hue = speed_data['Speed'])
```

```python
x_train = speed_data[['1 - Distance From Bump', 'Height of Bump']]
```

```python
y_train = speed_data['Speed']
```

```python
# Define and Fit the Model
model = Sequential()
model.add(Dense(output_dim=1, input_dim=2, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam')
model.fit(x_train, y_train, epochs=1000, verbose=2) #verbose: 0 for no logging, 1 for progress bar logging, 2 for one log line per epoch
```

```python
model.summary()
```

```python
# Make a Prediction
y_predict = model.predict_classes(x_train)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_train, y_predict)

sns.heatmap(cm, annot=True, fmt="d")
```

```python
# Visualising the Training Set Results
from matplotlib.colors import ListedColormap
x_set, y_set = x_train.values, y_train.values
X1, X2 = np.meshgrid(np.arange(start = x_set[:, 0].min() , stop = x_set[:, 0].max(), step = 0.01),
                     np.arange(start = x_set[:, 1].min(), stop = x_set[:, 1].max(), step = 0.01))
plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('blue', 'orange'))(i), label = j)
    
plt.title('Training Classification Results')
plt.xlabel('1-Distance from Bump')
plt.ylabel('Height of Bump')
plt.legend()
plt.show()
```
