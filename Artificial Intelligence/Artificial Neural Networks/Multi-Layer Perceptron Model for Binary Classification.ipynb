# Multi-Layer Perceptron Model for Binary Classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets

noisy_moons = datasets.make_moons(n_samples=1500, noise=.05) # Fabricate a challenging dataset
```

```python
x, y = noisy_moons
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets.samples_generator import make_blobs
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler # MinMaxScaler scales the values in range from 0 to 1

scalar = MinMaxScaler()

scalar.fit(x_train)
x_train_scaled = scalar.transform(x_train)

scalar.fit(x_test)
x_test_scaled = scalar.transform(x_test)
```

```python
plt.scatter(x_train_scaled[:, 0], x_train_scaled[:, 1], c=y_train)
```

```python
# Define and Fit the Model
model = Sequential()
model.add(Dense(20, input_dim=2, activation='relu')) # Input Layer  | Input dimension is required only for the first layer
model.add(Dense(20, activation='relu'))              # Hidden Layer | ReLU - Rectified Linear Units
model.add(Dense(20, activation='relu'))              # Hidden Layer 
model.add(Dense(20, activation='relu'))              # Hidden Layer 
model.add(Dense(output_dim=1, activation='sigmoid')) # Output Layer | Sigmoid activation function is used at the output so that the input can be classified using binary classification
model.compile(loss='binary_crossentropy', optimizer='adam')
model.fit(x_train_scaled, y_train, epochs=500, verbose=2)
```

```python
model.summary()
```

```python
# Make a Prediction
y_predict = model.predict_classes(x_test_scaled)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

cm = confusion_matrix(y_test, y_predict)

sns.heatmap(cm, annot=True, fmt="d")
```

```python
print(classification_report(y_test, y_predict))
```

```python
# Visualizing the Training Dataset Results
from matplotlib.colors import ListedColormap
x_grid, y_grid = x_train_scaled, y_train
X1, X2 = np.meshgrid(np.arange(start = x_grid[:, 0].min() - 1, stop = x_grid[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_grid[:, 1].min() - 1, stop = x_grid[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_grid)):
    plt.scatter(x_grid[y_grid == j, 0], x_grid[y_grid == j, 1],
                c = ListedColormap(('blue', 'orange'))(i), label = j)
    
plt.title('Training Dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```python
# Visualizing the Testing Dataset Results
from matplotlib.colors import ListedColormap
x_grid, y_grid = x_test_scaled, y_test
X1, X2 = np.meshgrid(np.arange(start = x_grid[:, 0].min() - 1, stop = x_grid[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_grid[:, 1].min() - 1, stop = x_grid[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, model.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_grid)):
    plt.scatter(x_grid[y_grid == j, 0], x_grid[y_grid == j, 1],
                c = ListedColormap(('blue', 'orange'))(i), label = j)
    
plt.title('Testing Dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```
