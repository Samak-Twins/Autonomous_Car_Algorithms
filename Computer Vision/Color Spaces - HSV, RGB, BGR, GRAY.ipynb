Convert from RGB to HSV

```python
image = cv2.imread('Self_Driving_Car.jpg')
cv2.imshow('Self Driving Car', image)

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV Image', hsv_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(hsv_image[:, :, 0])
plt.title('Hue Channel')
```

```python
plt.imshow(hsv_image[:, :, 1])
plt.title('Saturation Channel')
```

```python
plt.imshow(hsv_image[:, :, 2])
# add a title to the image
plt.title('Value Channel')
```

Split and merge R, G, B channels 

```python
image = cv2.imread('Self_Driving_Car.jpg')
B, G, R = cv2.split(image)
```

```python
B.shape
```

```python
R.shape
```

```python
G.shape
```

```python
# Split RGB channels of the original image
zeros = np.zeros(image.shape[:2], dtype = "uint8")

cv2.imshow("Red Channel", cv2.merge([zeros, zeros, R]))
cv2.imshow("Green Channel", cv2.merge([zeros, G, zeros]))
cv2.imshow("Blue Channel", cv2.merge([B, zeros, zeros]))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Merge RGB channels to re-create our original image
image_merged = cv2.merge([B, G, R]) 
cv2.imshow("Merged Image", image_merged) 

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Merge our original image while adding more color
image_extra_blue = cv2.merge([B+100, G, R])
image_extra_green = cv2.merge([B, G+100, R])
image_extra_red = cv2.merge([B, G, R+100])

cv2.imshow("Merged Image With Some Added Blue Channel", image_extra_blue)
cv2.imshow("Merged Image With Some Added Green Channel", image_extra_green)
cv2.imshow("Merged Image With Some Added Red Channel", image_extra_red)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
