# Find an Object using Template Matching

Load and display the image

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image_color = cv2.imread('Truck_1.jpg')
# image_color = mpimg.imread('Truck_1.jpg') # Import image as RGB instead of BGR 

cv2.imshow('Find a Truck', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width = image_color.shape[:2]
```

```python
height
```

```python
width
```

```python
plt.imshow(image_color)
```

Convert to Grayscale

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

```python
image_gray.shape
```

Load Template Image

```python
template = cv2.imread('Truck_Template.jpg',0)

cv2.imshow('Truck Template', template)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
template.shape
```

```python
w, h = template.shape[::-1]
w
```

```python
h
```

Perform Template Matching

```python
result = cv2.matchTemplate(image_gray, template, cv2.TM_CCOEFF_NORMED)
result.shape
```

```python
cv2.imshow('Matching Results', result)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Find the indices of the object to be found 
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
```

```python
max_loc
```

```python
# Draw a box around the object (if found)
top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(image_color, top_left, bottom_right, (0,255,0), 5)
```

```python
cv2.imshow('Truck Located', image_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
