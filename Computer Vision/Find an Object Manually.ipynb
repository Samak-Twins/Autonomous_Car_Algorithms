# Find an Object Manually

Load and display the image

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image_color = cv2.imread('Truck_1.jpg')
# image_color = mpimg.imread('Truck_1.jpg') # Import image as RGB instead of BGR 

cv2.imshow('Find a Truck', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```
