# Color Feature Detection - Histograms

Load and display the image

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
image_color = cv2.imread('Tesla_Car_Interior.jpg')

cv2.imshow('Original Image', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image_color)
```

```python
image_color.shape
```

```python
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histogram = cv2.calcHist([image_color], [i], None, [256], [0, 256])
    plt.plot(histogram, color = col)
    plt.xlim([0,256])
    
plt.show()
```
