# Cropping, Dilation and Erosion of Images

Load and display the image

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image = cv2.imread('Self_Driving_Car_2.jpg')
#image = mpimg.imread('Self_Driving_Car_2.jpg') # Import image as RGB instead of BGR

cv2.imshow('Original Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width = image.shape[:2]
```

```python
height
```

```python
width
```

Perform Cropping

```python
# Top left coordinates of the desired cropped area (w0, h0)
w0 = int(width * 0.5)
h0 = int(height * 0.5) 
```

```python
# Bottom right coordinates of the desired cropped area (w1, h1)
h1 = int(height * 1)
w1 = int(width * 1)
```

```python
Image_cropped = image[h0:h1 , w0:w1]

cv2.imshow("Cropped Image", Image_cropped) 
cv2.waitKey() 
cv2.destroyAllWindows()
```

```python
# If you want to see the actual distance translated
# Don't worry about the color, openCV uses BGR instead of RGB
plt.imshow(Image_cropped)
```

Perform Dilation and Erosion

```python
image = cv2.imread('Speed_Sign.jpg')
```

```python
kernel = np.ones((3,3), np.uint8)

image_erosion = cv2.erode(image, kernel, iterations=3)
image_dilation = cv2.dilate(image, kernel, iterations=3)

cv2.imshow('Input', image)
cv2.imshow('Erosion', image_erosion)
cv2.imshow('Dilation', image_dilation)

cv2.waitKey()
cv2.destroyAllWindows()
```
