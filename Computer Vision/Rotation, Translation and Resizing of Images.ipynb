# Rotation, Translation and Resizing of Images

Load and display the image

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image = cv2.imread('Self_Driving_Car_2.jpg')
cv2.imshow('Original Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width = image.shape[:2]
```

```python
height
```

```python
width
```

Perform Rotations

```python
M_rotation = cv2.getRotationMatrix2D((width/2, height/2), 45, 0.5) # Rotate around the center by 45 deg and scale 0.5 times
rotated_image = cv2.warpAffine(image, M_rotation, (width, height)) # Insert size of the output image

cv2.imshow('Rotated Image', rotated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

Perform Translations

```python
Translational_Matrix = np.float32([[1, 0, 300], 
                                  [0, 1, 50]])
```

```python
translated_image = cv2.warpAffine(image, Translational_Matrix, (width, height))
cv2.imshow('Translated Image', translated_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# Don't worry about the color, openCV uses BGR instead of RGB
plt.imshow(translated_image)
```

Perform Resizing

```python
resized_image = cv2.resize(image, None, fx=0.25, fy=0.25, interpolation = cv2.INTER_CUBIC) # Scale by 0.25 times in x & y directions
cv2.imshow('Resized Image', resized_image) 
cv2.waitKey()
cv2.destroyAllWindows()
```
