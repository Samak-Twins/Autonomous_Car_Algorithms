# Region of Interest (ROI) Masking

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image_color = cv2.imread('Road_Lane.jpg')
#image_color = mpimg.imread('Road_Lane.jpg') # Import image as RGB instead of BGR

cv2.imshow('Original Image', image_color)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
height, width = image_color.shape[:2]
```

Convert to Grayscale

```python
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap = 'gray')
```

```python
image_gray.shape
```

Perform Region of Interest Masking

```python
# Select points of the region of interest (ROI)
ROI = np.array([[(0, height),(320, 300), (500, 300), (width, height)]], dtype=np.int32)    

# Define a blank image with all zeros (i.e. black) 
blank = np.zeros_like(image_gray)   
```

```python
blank.shape
```

```python
# Fill the Region of interest with white color (i.e. 255)
mask = cv2.fillPoly(blank, ROI, 255)
    
# Perform bitwise AND operation to select only the region of interest
masked_image = cv2.bitwise_and(image_gray, mask)
```

```python
masked_image.shape
```

```python
plt.imshow(masked_image, cmap = 'gray')
```
