# Perspective Transform on Images

Load and display the image

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
from matplotlib import pyplot as plt
%matplotlib inline

image = cv2.imread('Speed_Sign.jpg')
cv2.imshow('Original Image', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image = mpimg.imread('Speed_Sign.jpg')
plt.imshow(image)
```

```python
height, width = image.shape[:2]
```

```python
# First entry: top-left,
# Second entry: top-right
# Third: bottom-right
# Fourth: bottom-left

# Cordinates of the 4 corners of the original image
Source_points = np.float32([[200,60], [450,150], [520, 500], [170,470] ])

# Cordinates of the 4 corners of the desired output
Destination_points = np.float32([[0,0], [width,0], [width,height], [0,height]])
```

```python
# Use the two sets of four points to compute the Perspective Transformation matrix, M    
M = cv2.getPerspectiveTransform(Source_points, Destination_points)
 
warped = cv2.warpPerspective(image, M, (width, height))
 
cv2.imshow('Warped Image', warped)
cv2.waitKey()
cv2.destroyAllWindows()
```
